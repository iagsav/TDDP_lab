## Групировка данных

Группировка при помощи метода groupby обьединять данные в группы по одному или нескольким признакам и   
считать по ним общую статистику

```python
pass_data.groupby('sex') # разбиение всех обьетов на 2 группы по полу
```

```python
pass_data.groupby('sex')['Pclass'].value_counts()
```

```python
pass_data.groupby('Pclass')['Fare'].describe()
```

```python
pass_data.groupby('sex')['Age'].mean() # средний возраст для пассажиров каждого из полов
```

Как известно , в превую очередь спасали женщин и детей в порядке повышения номера класса, в связи с этим доля   
выживших женщин, вероятно, будет выше, чем доля выживших мужчин. Проверим:

```python
pass_data.groupby('sex')['Survived'].mean()
```

Аналогично для пассажиров различных классов

```python
pass_data.groupby('Pclass')['Survived'].mean()
```

Рассмотренные выше статистические данные могут быль расссмотренны и в виде стандартного DataFrame

```python
pass_data.groupby('sex', as_index = False)['Survived'].mean()
```
